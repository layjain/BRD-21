```python
def date_within_range(date, start_date, end_date):
    date = datetime.datetime.strptime(date, "%Y-%m-%d")
    start_date = datetime.datetime.strptime(start_date, "%Y-%m-%d")
    end_date = datetime.datetime.strptime(end_date, "%Y-%m-%d")
    return start_date <= date <= end_date
```

```python
for i, row in df_response.iterrows():
    country = row["location"]
    measure = row["measure"]
    start_date = row["start_date"]
    end_date = "2021-03-07" if (type(row["end_date"] == float) and pd.isna(row["end_date"])) else row["end_date"]
    for idx, row_2 in df.iterrows():
        val = False
        curr_date = row_2["date"]
        if country == row_2["location"] and date_within_range(curr_date, start_date, end_date):
            df.at[idx, measure] = 1
```
